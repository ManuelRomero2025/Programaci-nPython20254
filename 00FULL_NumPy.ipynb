# **¿Qué es NumPy?**

image.png

NumPy (Numerical Python) es una biblioteca fundamental de Python para computación científica, que proporciona:

- Un objeto de arreglo multidimensional eficiente: ndarray.

- Operaciones vectorizadas sobre arrays (más rápidas que los bucles).

- Funciones matemáticas, estadísticas, lógicas, de álgebra lineal, Fourier, y más.

***Fue creado originalmente por Travis Oliphant en 2005 como una evolución de Numeric y Numarray.***

# **¿Por qué no basta solo Python?**

Python puro tiene listas (list) y bucles for, pero:

| Python puro                     | NumPy                                               |
| ------------------------------- | --------------------------------------------------- |
| Listas dinámicas y lentas       | Arreglos fijos y eficientes (`ndarray`)             |
| Sin operaciones vectorizadas    | Soporta cálculos vectorizados (`+`, `*`, `np.sum`)  |
| Menor control de tipos de datos | Control total con `dtype` (como `int64`, `float32`) |
| Ineficiente con grandes datos   | Rápido y optimizado en C internamente               |

## **¿Por qué es importante esto en Python y NumPy?**

En Python puro, los enteros (int) son de precisión arbitraria, es decir, pueden crecer según la memoria disponible. Pero en NumPy, se usa una representación más controlada para:

* Optimizar el uso de memoria

* Mejorar el rendimiento computacional

* Controlar el tipo y tamaño de datos en ciencia de datos, imágenes, redes neuronales, etc.

* Interoperabilidad con datos binarios, sensores, hardware o estructuras de C/C++

| Uso Real                           | Tipo sugerido     | Justificación                                                |
| ---------------------------------- | ----------------- | ------------------------------------------------------------ |
| Imagen en escala de grises         | `uint8`           | Valores de 0 a 255 (8 bits) para intensidad de píxeles       |

| Datos financieros grandes          | `int64`           | Rango suficiente para grandes montos y cálculos precisos     |
| Etiquetas de clasificación         | `int8`/`uint8`    | Clasificaciones que van de -1 a 1 o 0 a 10 (datos compactos) |
| Datos meteorológicos (temperatura) | `int16`           | Permite valores negativos y positivos en buen rango          |


## **Números (clases int y float)**

En NumPy, los tipos numéricos básicos como int (enteros) y float (decimales) se manejan de forma especializada para aprovechar al máximo el rendimiento y la memoria.


**¿Qué significa esto?**

Cuando se crea un np.array, NumPy infiere automáticamente el tipo de dato (dtype) según los valores incluidos, como:

* int32, int64 → enteros

* float32, float64 → números con decimales






```python
# Array con enteros (int)
import numpy as np

enteros = np.array([1, 2, 3, 4])
print("Array entero:", enteros)
print("Tipo de datos:", enteros.dtype)
```

```python
# Array con números decimales (float)
decimales = np.array([1.5, 2.0, 3.25])
print("Array decimal:", decimales)
print("Tipo de datos:", decimales.dtype)
```

```python
# Forzar tipo de datos con dtype
forzado_float32 = np.array([1, 2, 3], dtype=np.float32)
print("Array forzado a float32:", forzado_float32)
print("Tipo:", forzado_float32.dtype)

forzado_int8 = np.array([10, 20, 30], dtype=np.int8)
print("Array forzado a int8:", forzado_int8)
print("Tipo:", forzado_int8.dtype)
```

NumPy permite especificar el tipo de dato con dtype=np.int32, np.float64, etc.

Elegir el tipo adecuado es clave para:

* Ahorrar memoria

* Aumentar velocidad

* Evitar errores por desbordamientos o pérdidas de precisión

## Listas (list) + NumPy
Las listas son colecciones ordenadas, mutables, que pueden contener elementos de diferentes tipos. Una lista es una secuencias ordenadas de objetos de distintos tipos.

Se construyen poniendo los elementos entre corchetes [ ] separados por comas.

Se caracterizan por:

* Tienen orden.
* Pueden contener elementos de distintos tipos.
* Son mutables, es decir, pueden alterarse durante la ejecución de un programa.


```python
import numpy as np

#Usa corchetes [ ].
#Ordenada: conserva el orden de los elementos.
#Mutable: se puede modificar (agregar, eliminar o cambiar elementos)

# Lista con enteros
lista = [1, 2, 3, 4, 5]

# Convertir la lista en un array de NumPy
array_desde_lista = np.array(lista)

print("Lista:", lista)
print("Array desde lista:", array_desde_lista)
```

```python
matriz = [[1, 2, 3],
          [4, 5, 6]]
A = np.array(matriz)
print("Matriz:\n", A)
```

```python
A[0, 1] = 100
print("Matriz modificada:\n", A)
```

```python
[1, 2, 3] + [4, 5, 6]
```

```python
np.array([1,2,3]) + np.array([4,5,6])
```

## Tuplas (tuple) + NumPy

Una tupla es una secuencias ordenadas de objetos de distintos tipos. Se construyen poniendo los elementos entre corchetes ( ) separados por comas.

Se caracterizan por:

* Tienen orden.
* Pueden contener elementos de distintos tipos.
* Son inmutables, es decir, no pueden alterarse durante la ejecución de un programa. Inmutable significa que no se puede modificar después de creada.

Las tuplas son colecciones ordenadas, inmutables, que también pueden contener datos heterogéneos.

```python
# Tupla con valores
#Los paréntesis ( ) indican que es una tupla.
#Los elementos están ordenados (tiene un índice).
#Es inmutable: no se pueden cambiar los valores después de crearla.

tupla = (10, 20, 30)

# Convertir a array de NumPy
array_desde_tupla = np.array(tupla)

print("Tupla:", tupla)
print("Array desde tupla:", array_desde_tupla)
```

```python
#tupla[0]=20 al ser tuple no permite cambios
```

Se usa una tupla cuando se necesita guardar varios valores juntos y no se desea que cambien. A diferencia de una lista ([ ]), la tupla (( )) protege esos valores de modificaciones accidentales.

| Situación                              | Ejemplo                         | Explicación                                 |
| -------------------------------------- | ------------------------------- | ------------------------------------------- |
| Agrupar varios valores constantes      | `coordenada = (3, 5)`           | No se cambiará durante el programa          |
| Devolver varios valores de una función | `return (x, y)`                 | Retorna más de un dato                      |
| Iterar por pares o grupos fijos        | `for (a,b) in lista_de_tuplas:` | Desempaqueta elementos                      |
| Usar como clave en un diccionario      | `d[(1,2)] = "punto"`            | Las tuplas pueden ser claves, las listas no |


```python
A = np.array(((1, 2, 3), (4, 5, 6)))  # tupla de tuplas
print("Array A:\n", A)
print("Forma de A:", A.shape)
```

* La tupla no se puede modificar después.

* La lista puede alterarse antes de crear el array.

| Conversión    | Ejemplo          | Resultado   |
| ------------- | ---------------- | ----------- |
| Tupla → Lista | `list((1,2,3))`  | `[1, 2, 3]` |
| Lista → Tupla | `tuple([1,2,3])` | `(1, 2, 3)` |
| Array → Lista | `arr.tolist()`   | `[1, 2, 3]` |


 ## Diccionarios (dict) + NumPy
Los diccionarios asocian claves con valores. Aunque no se pueden convertir directamente en un array multidimensional, sí podemos extraer sus valores o convertir a estructuras útiles.

Un diccionario es una colección de pares formados por una clave y un valor asociado a la clave.

Se construyen poniendo los pares entre llaves { } separados por comas, y separando la clave del valor con dos puntos :.

Se caracterizan por:

* No tienen orden.
* Pueden contener elementos de distintos tipos.
* Son mutables, es decir, pueden alterarse durante la ejecución de un programa.
* Las claves son únicas, es decir, no pueden repetirse en un mismo diccionario, y pueden ser de cualquier tipo de datos inmutable.

```python
# Diccionario de constantes matemáticas
diccionario = {'pi': 3.1416, 'e': 2.718}

# Convertir los valores a un array de NumPy
array_valores = np.array(list(diccionario.values()))

print("Diccionario:", diccionario)
print("Array con valores del diccionario:", array_valores)
```

# **Operadores Lógicos con Números en NumPy**

En NumPy, se pueden aplicar operadores lógicos directamente sobre arrays numéricos. Estos devuelven un array de valores booleanos (True o False), resultado de comparar elemento por elemento.



```python
import numpy as np

a = np.array([5, 10, 15, 20])

# Comparaciones
print(a > 10)      # ¿Mayor que 10?
print(a <= 15)     # ¿Menor o igual a 15?
print(a == 10)     # ¿Igual a 10?
print(a != 15)     # ¿Distinto de 15?
```

```python
# Operadores combinados (&, |, ~)
# Filtrar los elementos entre 10 y 20 (ambos inclusive)
print((a >= 10) & (a <= 20))

# Elementos menores que 10 o mayores que 15
print((a < 10) | (a > 15))

# Negación (no mayores que 10)
print(~(a > 10))
```

En NumPy se usan &, |, ~ en lugar de and, or, not, ya que estos últimos no funcionan con arrays.

```python
# Filtrar con operadores lógicos

# Filtrar los valores del array que sean mayores que 10
mayores = a[a > 10]
print("Elementos mayores que 10:", mayores)

# Filtrar entre 10 y 15
rango = a[(a >= 10) & (a <= 15)]
print("Elementos entre 10 y 15:", rango)
```

```python
#  Comparación entre arrays
b = np.array([5, 12, 15, 25])

# Comparación elemento a elemento
print(a == b)
print(a < b)
```

Funciones lógicas útiles en NumPy

| Función          | Descripción                                        |
| ---------------- | -------------------------------------------------- |
| `np.all(cond)`   | Devuelve `True` si **todos** los elementos cumplen |
| `np.any(cond)`   | Devuelve `True` si **algún** elemento cumple       |
| `np.where(cond)` | Devuelve los índices donde la condición es `True`  |


```python
print(np.all(a > 0))          # True
print(np.any(a > 18))         # True
print(np.where(a > 10))       # (array([2, 3]),)
```

## Acceso a los elementos de una cadena

NumPy no trata directamente strings como cadenas mutables, pero puedes convertir una cadena en un array de caracteres:

| Índice Positivo | 0  | 1  | 2  | 3  | 4  | 5  |
| --------------- | -- | -- | -- | -- | -- | -- |
| Carácter        | P  | y  | t  | h  | o  | n  |
| Índice Negativo | -6 | -5 | -4 | -3 | -2 | -1 |

En Python, las cadenas (str) son secuencias indexadas.

Puedes acceder a cada carácter usando índices positivos (de izquierda a derecha) o negativos (de derecha a izquierda).

```python
import numpy as np

# Convertimos la cadena en un array de caracteres
cadena = "Python"
arr = np.array(list(cadena))

# Acceso por índice (igual que en listas)
print(arr[0])   # P
print(arr[3])   # h
print(arr[-1])  # n
```

```python
# Acceso con slicing (subcadenas)

# Obtener los tres primeros caracteres
print(arr[:3])  # ['P' 'y' 't']

# Últimos tres caracteres
print(arr[-3:]) # ['h' 'o' 'n']

# Subcadena intermedia
print(arr[1:5]) # ['y' 't' 'h' 'o']
```

```python
# Aplicación útil: Filtrar letras mayúsculas
# Detectar cuáles caracteres son mayúsculas
mayusculas = np.char.isupper(arr)
print(mayusculas)  # [ True False False False False False]

# Extraer sólo las mayúsculas
print(arr[mayusculas])  # ['P']
```
